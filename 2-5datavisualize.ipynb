```python
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
iris=datasets.load_iris()
iris_data=pd.DataFrame(iris['data'],columns=iris["feature_names"])
iris_target=pd.DataFrame(iris["target"], columns=["target"])
data=pd.concat([iris_data,iris_target],axis=1)
data
```

### specifying the colors of different iris flowers

```python
colors={0: "r", 1:"g", 2: "b"}
dot_colors=data["target"].map(colors)
data.plot.scatter(x="sepal length (cm)", y="sepal width (cm)", c=dot_colors)
```

### adding some labels in the graph

```python
iris["target_names"]
```

```python
ax1=data[data["target"]==0].plot.scatter(x="sepal length (cm)", y="sepal width (cm)", c="r", label="setosa")
ax2=data[data["target"]==1].plot.scatter(x="sepal length (cm)", y="sepal width (cm)", c="g", label="versicolor", ax=ax1, alpha=0.5)
data[data["target"]==2].plot.scatter(x="sepal length (cm)", y="sepal width (cm)", c="b", label="virginica", ax=ax2 ,alpha=0.5)
```
