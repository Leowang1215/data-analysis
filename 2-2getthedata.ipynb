```python
import requests
import pandas as pd
import json
```

```python
metro_nearby_bus=requests.get('https://data.taipei/api/v1/dataset/a504e332-9f23-4649-82df-23a113581fa8?scope=resourceAquire')
```

```python
metro_nearby_bus.text 
```

```python
metro_nearby_bus.json() #把json格式轉換為python的dict
```

```python
data=metro_nearby_bus.json()["result"]["results"]
```

```python
pd.DataFrame(data)
```

```python
busname=[]
for item in data: #他每一次會從data裡取東西出來放進item
    busname.append(item["公車名稱"]) #把item裡的資料叫做"公車名稱"的值抽出來，放進去busname
print(busname)
```
